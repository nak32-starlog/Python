```python
class DigitalWallet:
    def __init__(self, balance, pin_code):
        self.__balance = max(0, balance) 
        self.__pin_code = pin_code

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Successfully deposited ${amount}")
        else:
            print("Deposit amount must be positive.")

    def get_balance(self, pin):
        if pin == self.__pin_code:
            print(f"Your Balance: {self.__balance}")
            
        else:
            print("Access Denied")

    def withdrawal(self, amount, pin):
    
        if pin != self.__pin_code:
            print("Access Denied: Incorrect PIN")
            return

        if amount > self.__balance:
            print(f"Insufficient funds! Current balance: ${self.__balance}")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining: ${self.__balance}")

wallet = DigitalWallet(100, 1111)
wallet.get_balance(1111)
wallet.get_balance(19999)
wallet.withdrawal(500, 1111)  
wallet.withdrawal(50, 9999)  
wallet.withdrawal(50, 1111) 
wallet.get_balance(1111)
```

```python
class VirtualPet:
    def __init__ (self, name):
        self.name = name
        self.hunger = 5
        self.energy = 10
        self.happiness = 5
        
    def feed(self):
        self.hunger -= 1
        print(f"{self.name}'s Hunger: {self.hunger}")
            
    def play(self):
        self.happiness += 2
        self.energy -= 2
        print(f"{self.name}'s Energy: {self.energy}")
        print(f"{self.name}'s Happiness: {self.happiness}")
        
    def sleep(self):
        self.energy = 10
        print(f"{self.name}'s Energy: {self.energy}")
        
pet1 = VirtualPet("Bubbles")
print(pet1.name)
pet1.feed()
pet1.play()
pet1.sleep()
```

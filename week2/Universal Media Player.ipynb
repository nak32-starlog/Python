```python
class Media:
    def play(self):
        pass
        
class Song(Media):
    def play(self):
        print("Playing audio...")

class Movie(Media):
    def play(self):
        print("Playing video at 1080p...")

class Podcast(Media):
    def play(self):
        print("Playing podcast")

def play_all(media_list):
    for item in media_list:
        item.play()
        
media_items = [Song(), Movie(), Podcast()]
play_all(media_items)
```

# Number Guessing Game
#### The Goal: The program randomly selects a number between 1 and 100. The user guesses the number, and the program replies with "Too high!", "Too low!", or "You got it!". The game loops until the user guesses correctly.
#### Concepts Tested: while loops, if/elif/else statements, comparison operators, importing the basic random module.

```python
import random 

secret_number = random.randint(1,100);
guess = 0;

while guess != secret_number: 
    guess = int(input("Enter number between 1 to 100: "));

    if guess > secret_number:
        print("Too High!");
    elif guess < secret_number:
        print("Too Low!");
    else:
        print(f"You got it! the number was {secret_number}");
```

# Rock, Paper, Scissors:
#### The Goal: A continuous loop where the user plays Rock, Paper, Scissors against the computer. Keep track of the score until the user types "quit".
#### Concepts Tested: Infinite loops with break conditions, complex if/elif logic, string normalization (e.g., using .lower()).

```python
import random

user_score = 0
computer_score = 0
options = ["rock", "paper", "scissors"]

while True:
    user_input = str(input("Rock, Paper, Scissors or 'quit' to exit :")).lower();

    if user_input == "quit":
        break;
    elif user_input not in options:
        print("Invalid choice, please try again");
        continue;

    computer_pick = random.choice(options)
    print(f"Computer Picked {computer_pick}")

    if user_input == computer_pick:
        print("It's tie")
    elif (user_input == "rock" and computer_pick == "scissors") or\
         (user_input == "scissors" and computer_pick == "paper") or\
         (user_input == "paper" and computer_pick == "rock"):
        print("You won!")
        user_score +=1
    else:
        print("You lose!")
        computer_score +=1
        
    print(f"Score -> You: {user_score} | Computer: {computer_score}\n")
    
```

# The Mental Math Challenger
#### The Goal: The program generates two random numbers between 1 and 12 and asks the user to multiply them together (e.g., "What is 7 x 8?"). If the user gets it right, they get a point, and the program asks another question. If they get it wrong, the program tells them the correct answer and ends the game, displaying their final streak/score.
#### Concepts Tested: Importing the random module, tracking score variables, while loops dependent on a boolean condition (e.g., while is_correct:), if/else statements, and type casting (converting the user's answer to an int to compare it to the actual math result).

```python
import random

score = 0
is_correct = True

print("---------- Welcome to the Mental Math Challenger! -----------")
print("Get as many right as you can. One mistake and it's Game Over!")
print("-------------------------------------------------------------")

while is_correct:
    num1 = random.randint(1,12)
    num2 = random.randint(1,12)
    actual_answer = num1 * num2
    
    user_guess = int(input(f"What is {num1} x {num2}?: "))
    
    
    if user_guess == actual_answer:
        score += 1
        print(f"Correct! Current Streak: {score}\n")
    else:
        print(f"\nWrong! The correct answer was {actual_answer}.")
        is_correct = False

print("-------------------------------------------------------------")
print(f"           GAME OVER! Your final score: {score}")
print("          Can you beat your streak next time?")
```

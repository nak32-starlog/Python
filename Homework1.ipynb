# Mad Libs Game with Python

```python
print("*" * 50);
print(" "*14 + "WELCOME TO MAD LIBS");
print("*" * 50);
print("Answer these questions below, then enjoy");
print("the sily story you created\n");

adj1 = str(input("Enter an adjective (e.g happy, loud, red): "));
adj2 = str(input("Enter another adjective (e.g cooked, goofy, cringe): "));

noun1 = str(input("Enter a noun (e.g dog, table, chair): "));
noun2 = str(input("Enter another noun (e.g car, river, bridge): "));

verb1 = str(input("Enter a verb (e.g fly, ride, rizz): "));
verb2 = str(input("Enter another verb (e.g beat, cook, destroy): "));

place = str(input("Enter a place (e.g office, library, market): "));
animal = str(input("Enter an animal (e.g cat, fish, capybara): "));
dance = str(input("Enter a dance type (e.g beksloy, kalok, rom vung): "));
celebrity_name = str(input("Enter a celebrity name (e.g Sinn Sisamouth, Preap Sovath, Vannda): "));
clothing = str(input("Enter a clothing type (e.g shirt, trouser, coat): "));
meme_catchphrase = str(input("Enter a meme catchphrase (e.g it's not that deep, caught in 4k, must've been the wind): "));
number = int(input("Enter number (e.g 5, 67, 1234): "));

story = f"""
{'─' * 75}
 YOUR MAD LIBS STORY
{'─' * 75}

I woke up feeling absolutely {adj1}.
I checked my {noun1} and realized I was late for {place}.
I grabbed my {noun2} and sprinted out the door like a {animal}.

When I arrived, everyone was {verb1} because {celebrity_name} was there doing the {dance}. 
I tried to join in, but I accidentally {verb2} my {clothing}.

The principal gave me {number} days of detention, which was totally {adj2}. 
I just looked at the camera and said, {meme_catchphrase}.

{'─' * 75}
"""
print(story);
```

# Resturant Tips Calculator 

```python
bill = float(input("What was the total bill amount? $"))
tip_percentage = float(input("What percentage tip would you like to give? "))

tip_amount = bill * (tip_percentage / 100)
total = bill + tip_amount

print(f"Tip Amount: ${tip_amount:.2f}")
print(f"Grand Total: ${total:.2f}")
```

# The Road Trip Cost Estimator

```python
distance = float(input("Enter your total trip distance (km): "));
kpl = float(input("Enter car average fuel efficiency ( kilometer per liter ): "));
fuel_price = float(input("Enter the current fuel price per liters: "));

fuel_needed = distance / kpl;
total_cost = fuel_needed * fuel_price;

print("\n--- Road Trip Summary ---")
print(f"Total Fuel Needed: {fuel_needed:.2f} Liters")
print(f"Total Estimated Cost: ${total_cost:.2f}")
```

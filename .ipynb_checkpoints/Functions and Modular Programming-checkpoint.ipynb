# Modular Unit Converter:
#### The Goal: A program that converts temperatures (Celsius to Fahrenheit), distances (Miles to Kilometers), and weights(Grams to Kilograms).
#### The Catch: Every single conversion must be its own independent function (e.g., def c_to_f(celsius):). There should be a main menu function that calls the others based on user choice.
#### Concepts Tested: Defining functions, passing arguments, return statements, variable scope.

```python
print("---Conversion Program---")
print("1.Temp 2.Distance 3.Weight 4.Exit")

def main():
    while True:
        user_input = input(" Enter Your Choice:")

        if user_input == "1":
            celcius = float(input("Enter Temperature as Calcius: "))
            result = cel_to_fahr(celcius)
            print(f"Celcius to Fahrenheit Conversion: {result:.2f}")
            
        elif user_input == "2":
            miles = float(input("Enter Distance as Miles: "))
            result = mile_to_km(miles)
            print(f"Miles to Kilometers Conversion: {result:.2f}")
            
        elif user_input == "3":
            gram = float(input("Enter Weight as Gram"))
            result = g_to_kg(gram)
            print(f"Grams to Kilograms Conversion: {result:.2f}")
            
        elif user_input == "4":
            print("Good Bye!")
            break
            
        else:
            print("Invalid choice, Please try again")

def cel_to_fahr(celcius):
    fahrenheit = celcius * 8 + 32
    return fahrenheit

def mile_to_km(miles):
    km = miles * 1.609
    return km

def g_to_kg(gram):
    kg = gram * 1000
    return kg

main()
```

# Password Generator:
#### The Goal: A function that takes an integer (length) and boolean flags (use_numbers, use_symbols) and returns a randomized, secure password.
#### Concepts Tested: Default arguments, keyword arguments, combining lists/strings, returning data.

```python
import random as rd
import string as s
    
def generate_password(length=12, use_numbers=False, use_symbols=False):
    
    pool = s.ascii_letters
    if use_numbers == True: 
        pool += s.digits
    if use_symbols == True: 
        pool += s.punctuation
    result_list = rd.choices(pool, k=length)
    password = "".join(result_list)

    #extra
    if length < 1:
        return("Length must be at least 1.")
    if not pool: 
        return("No Characters Selected")
        
    safety = "Weak"
    if len(password) >= 8 and use_numbers:
        safety = "Medium"
    if len(password) >= 12 and use_numbers and use_symbols:
        safety = "Strong"
        
    return password, safety

print(generate_password(length=0))
print(generate_password(length=10, use_numbers=True, use_symbols=False))
print(generate_password(length=12, use_numbers=True, use_symbols=True))
print(generate_password())
```

# The Text Analyzer Suite
#### The Goal: Ask the user to input a long sentence or paragraph. Build a program that outputs statistics about their text.
#### The Catch: You must use separate functions for every analysis: count_words(text), count_vowels(text), and find_longest_word(text). Each function must take the text as an argument and return the result to be printed by the main program.
#### Concepts Tested: String manipulation, passing strings as arguments, returning specific values (integers and strings) from functions, and keeping logic strictly separated.

```python
import string as s

def main():
    user_text = str(input("Enter your paragraph: "))
    print(f"Total Words: {count_words(user_text)}")
    print(f"Total Vowels: {count_vowels(user_text)}")
    print(f"Longest Word: {find_longest_words(user_text)}")
    
def count_words(text):
    words = text.split()
    return len(words)
    
def count_vowels(text):
    vowels = "aeiou"
    vowel_counts = 0
    for char in text.lower():
        if char in vowels:
            vowel_counts +=1
    return vowel_counts  
    
def find_longest_words(text):
    words = text.split()
    longest_words = ""
    for word in words:
        if len(word) > len(longest_words):
            longest_words = word
    return longest_words
    
main()
```

# The Virtual ATM (Banking Simulator)
#### The Goal: Simulate a bank account where a user can deposit, withdraw, or check their balance.
#### The Catch: Create specific functions like deposit(balance, amount) and withdraw(balance, amount). These functions must return the newly calculated balance to overwrite the old balance in the main program loop. The withdraw function must also include logic to prevent overdrafting (returning an error message and the unmodified balance if they ask for too much).
#### Concepts Tested: Passing variables as arguments, updating state using return statements, handling edge cases within functions, and understanding variable scope (realizing that a variable modified inside a function doesn't change outside unless returned).

```python
def main():
    balance = 0
    print("1.Balance_Check 2.Desposit 3.Withdraw 4.Exit")
    while True:
        user_input = input("Enter your choice: ")
        
        if user_input == "1":
            check_balance(balance)  
            
        elif user_input == "2":
            try:
                amount = float(input("Enter deposit amount: "))
            except ValueError:
                print("Please enter amount as numeric!")
                continue
            balance = deposit(balance, amount)
            
        elif user_input == "3":
            try:
                amount = float(input("Enter withdraw amount: "))
            except ValueError:
                print("Please enter amount as numeric!")
                continue
            balance = withdraw(balance, amount)
            
        elif user_input == "4":
            print("Good Bye!")
            break
        else:
            print("Invalid choice, Please try again")
            continue

def check_balance(balance):
    print(f"Your current balance is: ${balance:.2f}")
    
def deposit(balance, amount):
    if amount > 0:
        balance += amount
        print(f"Successful Desposit {amount:.2f}")
    else:
        print("Amount can't be negative or 0 when deposit")
    return balance
        
def withdraw(balance, amount):
    if amount <= 0:
        print("Amount can't be negative or 0 when withdraw")
    elif amount <= balance:
        balance -= amount
        print(f"Successful Withdraw {amount:.2f}")
    else:
        print("Not enough balance!")
    return balance
    
main()
```

# Memory-Based To-Do List:
#### The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
#### Concepts Tested: List methods (.append(), .pop()), indexing, iterating through lists with for loops.

```python
tasks = [];

while True:
    
    user_input = str(input("type ADD, REMOVE, VIEW, or QUIT to exit :")).lower().strip()
    
    if user_input == "add":
        task_add = str(input("What task do you want to add: "))
        tasks.append(task_add)
        
    elif user_input == "remove":
         #extra
         if not tasks:
             print("There's nothing to delete!")
         else:
             task_remove = int(input("What task number do you want to remove: "))
             
             if 1 <= task_remove <= len(tasks):
                 tasks.pop(task_remove - 1)
                 print("Task removed successfully.")
             else:
                 print("Invalid task number! Please choose a number from the list.")

    elif user_input == "view":
        #extra
        if len(tasks) == 0:
            print("No tasks available")
        else:
            for index, task in enumerate(tasks, start=1):
                print(f"{index}. {task}")      
        
    elif user_input == "quit":
        break
        
    else: 
        print("Invalid choice, please try again");
        continue;
        
```

# The Coffee Shop Inventory Manager
#### The Goal: Create a dictionary pre-filled with cafe supplies and their quantities (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks the user if they want to "sell" or "restock" an item. Update the dictionary values based on their input and print the current inventory levels. Prevent the user from selling more items than they have in stock!
#### Concepts Tested: Accessing and updating dictionary values, checking if a key exists using the in operator, combining if/elif logic with dictionary lookups, and managing state.

```python
inventory = {
    "beans": 50, 
    "cups": 100, 
    "syrup": 15
}
while True:
    
    user_input = input("Type: Sell, Restock, View, Exit: ").lower().strip()
    
    if user_input == "view":
        for item, quantity in inventory.items():
            print(f"{item.capitalize()}: {quantity} units")

    elif user_input == "sell":
        item_name = str(input("Which item do you want to sell?:")).lower().strip()
        if item_name in inventory:
            amount = int(input("How many item do you want to sell?"))
            if amount > inventory[item_name]:
                 print(f"Not enough stock! Only {inventory[item_name]} left.")
            else:
                inventory[item_name] -= amount
                print(f"Successfully sold {amount} {item_name}. Remaining stock: {inventory[item_name]}")
        else:
            print("Items name is not available!")
            
    elif user_input == "restock":
        item_name = str(input("Which item do you want to restock?:")).lower().strip()
        
        if item_name in inventory:
            amount = int(input("How many item do you want to restock?"))
            inventory[item_name] += amount
            print(f"Successfully restocked {amount} {item_name}. New total: {inventory[item_name]}")
        else:
            print("Items name is not available!")
            
    elif user_input == "exit":
        print("Successful exit the system!")
        break
```
